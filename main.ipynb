```python
from neo4j import GraphDatabase

# Connection details
URI = 'neo4j://127.0.0.1:7687'
AUTH = ("neo4j", "password") 

def test_connection():
    try:
        # Create the driver
        driver = GraphDatabase.driver(URI, auth=AUTH)
        # Check if it connects
        driver.verify_connectivity()
        print("Everything is set up correctly! Python can see Neo4j.")
        driver.close()
    except Exception as e:
        print(f"The code is right, but the database isn't responding: {e}")

if __name__ == "__main__":
    test_connection()
```

```python
from neo4j import GraphDatabase

URI = "neo4j://127.0.0.1:7687"
AUTH = ("neo4j", "password")

driver = GraphDatabase.driver(URI, auth=AUTH)

def run_query(query):
    with driver.session(database="neoj") as session:
        result = session.run(query)
        return [record.data() for record in result]

if __name__ == "__main__":
    output = run_query("MATCH (n) RETURN n LIMIT 5")
    print(output)
    driver.close()
```

```python
from neo4j import GraphDatabase

URI = "neo4j://127.0.0.1:7687"
AUTH = ("neo4j", "password")   # change password

driver = GraphDatabase.driver(URI, auth=AUTH)

def create_movie(tx):
    query = """
MATCH (p:Person)-[:STUDIES]->(f:Field)
RETURN p.name AS person, f.name AS field
"""
    tx.run(query)

with driver.session(database="neo4j") as session:
    session.execute_write(create_movie)

driver.close()

print("Data created from Python successfully!")
```

```python
query = """
MATCH (p:Person)-[:STUDIES]->(f:Field)
RETURN p.name AS person, f.name AS field
"""

from neo4j import GraphDatabase

URI = "neo4j://127.0.0.1:7687"
AUTH = ("neo4j", "password")

driver = GraphDatabase.driver(URI, auth=AUTH)

with driver.session(database="neo4j") as session:
    result = session.run(query)

    for record in result:
        print(record["person"], "studies", record["field"])

driver.close()
```
